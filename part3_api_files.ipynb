```python
# ============================================================
# Assignment Part 3: File I/O, APIs & Exception Handling
# Theme: Product Explorer & Error-Resilient Logger
# ============================================================

import requests  # for API calls
from datetime import datetime  # for getting the current date and time

# ==============================================================
# TASK 1 — File Read & Write Basics
# ==============================================================

print("*" * 50)
print("TASK 1 — File Read & Write Basics")
print("*" * 50)

# ---- PART A: Write to file ----

notes = [
    "Topic 1: Variables store data. Python is dynamically typed.",
    "Topic 2: Lists are ordered and mutable.",
    "Topic 3: Dictionaries store key-value pairs.",
    "Topic 4: Loops automate repetitive tasks.",
    "Topic 5: Exception handling prevents crashes.",
]

with open("python_notes.txt", "w", encoding="utf-8") as f:
    for line in notes:
        f.write(line + "\n")
print("File written successfully.")

# Append two more lines
extra_lines = [
    "Topic 6: Functions help us organise and reuse code.",
    "Topic 7: Libraries extend Python's built-in capabilities.",
]

with open("python_notes.txt", "a", encoding="utf-8") as f:
    for line in extra_lines:
        f.write(line + "\n")
print("Lines appended.")

# ---- PART B: Read the file back ----
print("\n--- Reading python_notes.txt ---")

with open("python_notes.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

# Print each line with a number
for i, line in enumerate(lines, start=1):
    print(f"  {i}. {line.rstrip()}")

print(f"\nTotal lines in file: {len(lines)}")

# Keyword search
keyword = input("\nEnter a keyword to search in the notes: ").strip()
print(f"\nLines containing '{keyword}':")
found = False
for line in lines:
    # "Python" and "python" and "PYTHON" will all match "python"
    if keyword.lower() in line.lower():
        print(f"  → {line.rstrip()}")
        found = True

if not found:
    print(f"  No lines found containing '{keyword}'.")
```

```python
# ==============================================================
# TASK 2 — API Integration
# ==============================================================


print("\n" + "*" * 50)
print("TASK 2 — API Integration")
print("*" * 50)

BASE_URL = "https://dummyjson.com/products"

# ---- Step 1: Fetch 20 products and display them ----
print("\n--- Fetching 20 products ---")

try:
    response = requests.get(f"{BASE_URL}?limit=20", timeout=5)
    data = response.json()
    products = data["products"]

    # Print table header
    print(f"{'ID':<4} | {'Title':<32} | {'Category':<14} | {'Price':<9} | {'Rating'}")
    print("-" * 75)

    for p in products:
        print(f"{p['id']:<4} | {p['title']:<32} | {p['category']:<14} | ${p['price']:<8.2f} | {p['rating']}")

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

# ---- Step 2: Filter (rating >= 4.5) and Sort (price descending) ----
print("\n--- Products with rating >= 4.5, sorted by price (high to low) ---")

try:
    response = requests.get(f"{BASE_URL}?limit=20", timeout=5)
    products = response.json()["products"]

    high_rated = [p for p in products if p["rating"] >= 4.5]

    sorted_products = sorted(high_rated, key=lambda p: p["price"], reverse=True)

    print(f"{'ID':<4} | {'Title':<32} | {'Price':<9} | {'Rating'}")
    print("-" * 60)
    for p in sorted_products:
        print(f"{p['id']:<4} | {p['title']:<32} | ${p['price']:<8.2f} | {p['rating']}")

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

# ---- Step 3: Search by category (laptops) ----
print("\n--- Laptops category ---")

try:
    response = requests.get(f"{BASE_URL}/category/laptops", timeout=5)
    laptops = response.json()["products"]

    for laptop in laptops:
        print(f"  {laptop['title']:<35} ${laptop['price']:.2f}")

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

# ---- Step 4: POST request — create a product ----
print("\n--- Sending POST request to create a product ---")

new_product = {
    "title":       "My Custom Product",
    "price":       999,
    "category":    "electronics",
    "description": "A product I created via API"
}

try:
    response = requests.post(f"{BASE_URL}/add", json=new_product, timeout=5)
    print("Server response:")
    print(response.json())

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# ==============================================================
# TASK 3 — Exception Handling
# ==============================================================

print("\n" + "*" * 50)
print("TASK 3 — Exception Handling")
print("*" * 50)

# ---- Part A: Guarded Calculator ----
def safe_divide(a, b):
    """
    Divide a by b safely.
    Returns the result, or an error string if something goes wrong.
    """
    try:
        return a / b
    except ZeroDivisionError:
        # This triggers when b == 0
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print("\n--- Part A: safe_divide tests ---")
print(f"  safe_divide(10, 2)    → {safe_divide(10, 2)}")
print(f"  safe_divide(10, 0)    → {safe_divide(10, 0)}")
print(f"  safe_divide('ten', 2) → {safe_divide('ten', 2)}")

# ---- Part B: Guarded File Reader ----
def read_file_safe(filename):
    """
    Safely read a file.
    Returns the content if found; prints an error message if not.
    The 'finally' block always runs — success or failure.
    """
    try:
        with open(filename, "r", encoding="utf-8") as f:
            content = f.read()
        return content
    except FileNotFoundError:
        print(f"  Error: File '{filename}' not found.")
        return None
    finally:
        print(f"  File operation attempt complete.")

print("\n--- Part B: read_file_safe tests ---")
print("Reading 'python_notes.txt':")
content = read_file_safe("python_notes.txt")
if content:
    print(content[:100] + "...")

print("\nReading 'ghost_file.txt' (doesn't exist):")
read_file_safe("ghost_file.txt")

# ---- Part C: Robust API Calls (already covered in Task 2 with try/except blocks) ----
print("\n--- Part C: Robust API calls are included in Task 2 above ---")
print("   Every requests.get() and requests.post() is wrapped in try/except.")

# ---- Part D: Input Validation Loop ----
print("\n--- Part D: Product Lookup Loop ---")
print("(Type a product ID 1-100 or 'quit' to exit)")

while True:
    user_input = input("\nEnter a product ID (1-100) or 'quit': ").strip()

    # Exit condition
    if user_input.lower() == "quit":
        print("Exiting product lookup.")
        break

    # Validate: must be an integer
    try:
        product_id = int(user_input)
    except ValueError:
        print(f"  ⚠ '{user_input}' is not a valid integer. Please try again.")
        continue  # 'continue' skips the rest of this loop iteration and goes back to top

    # Validate: must be in range 1-100
    if not (1 <= product_id <= 100):
        print(f"  ⚠ ID must be between 1 and 100. Got {product_id}.")
        continue

    # Make the API call
    try:
        response = requests.get(f"{BASE_URL}/{product_id}", timeout=5)

        if response.status_code == 404:
            print(f"  Product ID {product_id} not found (404).")
        elif response.status_code == 200:
            product = response.json()
            print(f"  ✓ {product['title']} — ${product['price']:.2f}")
        else:
            print(f"  Unexpected status code: {response.status_code}")

    except requests.exceptions.ConnectionError:
        print("  Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("  Request timed out. Try again later.")
    except Exception as e:
        print(f"  Error: {e}")
```

```python
# ==============================================================
# TASK 4 — Logging to File
# ==============================================================

print("\n" + "*" * 50)
print("TASK 4 — Logging to File")
print("*" * 50)

LOG_FILE = "error_log.txt"

def log_error(location, error_type, message):
    """
    Append a timestamped error entry to the log file.
    - location:   where in the code the error happened (e.g. 'fetch_products')
    - error_type: what kind of error (e.g. 'ConnectionError')
    - message:    the error details
    """

    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_entry = f"[{timestamp}] ERROR in {location}: {error_type} — {message}\n"

    with open(LOG_FILE, "a", encoding="utf-8") as log:
        log.write(log_entry)

    print(f"  Logged: {log_entry.strip()}")

# ---- Trigger 1: ConnectionError from an unreachable URL ----
print("\n--- Triggering ConnectionError ---")
try:
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError as e:
    log_error("fetch_products", "ConnectionError", str(e)[:60])
except Exception as e:
    log_error("fetch_products", "UnexpectedError", str(e)[:60])

# ---- Trigger 2: HTTP 404 for a non-existent product ID ----
print("\n--- Triggering HTTP 404 for product ID 999 ---")
try:
    response = requests.get(f"{BASE_URL}/999", timeout=5)
    if response.status_code != 200:

        log_error("lookup_product", "HTTPError", f"{response.status_code} Not Found for product ID 999")
    else:
        print("  Product found (unexpected).")
except requests.exceptions.ConnectionError:
    log_error("lookup_product", "ConnectionError", "No internet connection")
except Exception as e:
    log_error("lookup_product", "UnexpectedError", str(e)[:60])

# ---- Print the full log file ----
print("\n--- Contents of error_log.txt ---")
try:
    with open(LOG_FILE, "r", encoding="utf-8") as log:
        print(log.read())
except FileNotFoundError:
    print("Log file not found — no errors were logged yet.")
```
